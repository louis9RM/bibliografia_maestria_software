```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Sequential
import matplotlib.pyplot as plt

# Configuraciones
batch_size = 64
noise_dim = 100  # Dimensión del vector de ruido
epochs = 10000  # Número de épocas
buffer_size = 60000  # Tamaño del dataset para shuffle

# Cargar y preparar el dataset MNIST
(x_train, _), (_, _) = tf.keras.datasets.mnist.load_data()
x_train = (x_train.astype('float32') - 127.5) / 127.5  # Normalizar entre -1 y 1
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)  # Redimensionar

# Crear el dataset de TensorFlow
dataset = tf.data.Dataset.from_tensor_slices(x_train).shuffle(buffer_size).batch(batch_size, drop_remainder=True)

# Función para construir el generador
def build_generator():
    model = Sequential([
        layers.Dense(256, input_dim=noise_dim, activation='relu'),
        layers.BatchNormalization(),
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dense(28 * 28 * 1, activation='tanh'),
        layers.Reshape((28, 28, 1))
    ])
    return model

# Función para construir el discriminador
def build_discriminator():
    model = Sequential([
        layers.Flatten(input_shape=(28, 28, 1)),
        layers.Dense(512, activation='relu'),
        layers.Dense(256, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

# Crear el generador y el discriminador
generator = build_generator()
discriminator = build_discriminator()
discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Congelar el discriminador para el modelo GAN
discriminator.trainable = False
gan_input = layers.Input(shape=(noise_dim,))
gan_output = discriminator(generator(gan_input))
gan = tf.keras.Model(gan_input, gan_output)
gan.compile(optimizer='adam', loss='binary_crossentropy')

# Función para generar ruido
def generate_noise(batch_size, noise_dim):
    return np.random.normal(0, 1, (batch_size, noise_dim))

# Función para mostrar imágenes generadas
def plot_images(generator, noise_dim, examples=16):
    noise = generate_noise(examples, noise_dim)
    generated_images = generator.predict(noise)
    generated_images = 0.5 * generated_images + 0.5  # Desnormalizar
    fig, axes = plt.subplots(4, 4, figsize=(4, 4))
    for i, ax in enumerate(axes.flatten()):
        ax.imshow(generated_images[i, :, :, 0], cmap='gray')
        ax.axis('off')
    plt.tight_layout()
    plt.show()

# Bucle de entrenamiento
for epoch in range(epochs):
    for real_images in dataset:  # Iterar sobre datos reales
        # Generar imágenes falsas
        noise = generate_noise(batch_size, noise_dim)
        fake_images = generator.predict(noise)

        # Crear etiquetas para discriminador
        real_labels = np.ones((batch_size, 1))  # Etiquetas para imágenes reales
        fake_labels = np.zeros((batch_size, 1))  # Etiquetas para imágenes falsas

        # Entrenar el discriminador
        d_loss_real = discriminator.train_on_batch(real_images, real_labels)
        d_loss_fake = discriminator.train_on_batch(fake_images, fake_labels)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Entrenar el generador
        noise = generate_noise(batch_size, noise_dim)  # Nuevo ruido para el generador
        misleading_labels = np.ones((batch_size, 1))  # El generador intenta engañar
        g_loss = gan.train_on_batch(noise, misleading_labels)

    # Mostrar progreso
    g_loss_value = g_loss[0] if isinstance(g_loss, list) else g_loss
    if epoch % 1000 == 0:
        print(f"Epoch {epoch} | D Loss: {d_loss[0]:.4f} | G Loss: {g_loss_value:.4f}")
        plot_images(generator, noise_dim)
```
