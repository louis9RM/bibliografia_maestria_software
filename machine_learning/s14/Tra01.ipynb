```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
# Cargar modelo preentrenado (GPT-2 en este caso)
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
# Frase inicial
input_text = "The ice cream is very"
```

```python
# Tokenizar la entrada
input_ids = tokenizer.encode(input_text, return_tensors="pt")
```

```python
# Generar predicción
output = model.generate(input_ids,
                        max_new_tokens=5,
                        num_return_sequences=1,
                        pad_token_id=tokenizer.eos_token_id,
                        temperature=0.7, #Controlar la aleatoriedad (más bajo = más determinista)
                        top_k=50, #Controlar la diversidad (50 palabras más importantes) (más alto = más diverso)
                        do_sample=True, #Activar la generación aleatoria
                        )
```

```python
# Decodificar la predicción
predicted_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Frase completa:", predicted_text)
```
