## ¿Cómo utilizar ImageNet?

Para utilizar ImageNet en la práctica, se pueden descargar conjuntos de datos y modelos pre-entrenados a través de bibliotecas de aprendizaje profundo como TensorFlow y PyTorch. Aquí hay un ejemplo simple de cómo cargar un modelo pre-entrenado de ImageNet en Keras.

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
# Cargar el modelo pre-entrenado ResNet50
model = ResNet50(weights='imagenet')
```

```python
# Cargar una imagen y preprocesarla
img_path = 'elefante.jpg'  # Ruta a la imagen
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
# Realizar la predicción
preds = model.predict(x)
```

```python
# Decodificar y mostrar los resultados
print('Predicho:', decode_predictions(preds, top=3)[0])
```

```python
img_path2 = 'golden.jpg'  # Ruta a la imagen
img2 = image.load_img(img_path2, target_size=(224, 224))
x2 = image.img_to_array(img2)
x2 = np.expand_dims(x2, axis=0)
x2 = preprocess_input(x2)
```

```python
# Realizar la predicción
preds2 = model.predict(x2)
```

```python
# Decodificar y mostrar los resultados
print('Predicho:', decode_predictions(preds2, top=5)[0])
```
