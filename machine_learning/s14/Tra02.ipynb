```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
# Cargar modelo preentrenado en español (GPT-2 español)
tokenizer = AutoTokenizer.from_pretrained("mrm8488/spanish-gpt2")
model = AutoModelForCausalLM.from_pretrained("mrm8488/spanish-gpt2")
```

```python
# Frase inicial
input_text = "El clima es muy"
```

```python
# Tokenizar la entrada
input_ids = tokenizer.encode(input_text, return_tensors="pt")
```

```python
# Generar predicción
output = model.generate(
    input_ids,
    max_new_tokens=10,  # Generar hasta 10 palabras adicionales
    num_return_sequences=1,  # Solo una secuencia de salida
    pad_token_id=tokenizer.eos_token_id,  # Configurar pad token
    temperature=0.8,  # Mantener cierta diversidad
    top_k=40,  # Limitar a las 40 palabras más probables
    repetition_penalty=1.2,  # Penalizar repeticiones
    do_sample=True  # Habilitar muestreo para generar palabras aleatorias
)
```

```python
# Decodificar la predicción
predicted_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Frase completa:", predicted_text)
```
